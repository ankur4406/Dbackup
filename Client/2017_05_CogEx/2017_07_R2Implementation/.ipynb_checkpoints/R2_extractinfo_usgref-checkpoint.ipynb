```python
# Setting up by importing all the relevant libraries
import os
import cv2
import matplotlib.pyplot as plt
from scipy import misc
```

```python
# User Inputs
dir_name = 'C:\Users\\ankuarora\Desktop\Client\\2017_05_CogEx\\2017_07_R2Implementation\Formsforinput'
base_filename = 'CMS1500_FrasierClean.png'
ref_start_filename = 'start_marker.png'
ref_end_filename = 'end_marker.png'
```

```python
# Loading the incoming CMS1500 Form in the form of an array
img_form = cv2.imread(os.path.join(dir_name, base_filename))

# Converting the BGR image into a Grayscale image
img_form_gray = cv2.cvtColor(img_form, cv2.COLOR_BGR2GRAY)

# Checking details about the loaded image
print(img_form_gray.shape, img_form_gray.size, img_form_gray.dtype)
```

```python
# Loading the reference markers
ref_start = cv2.imread(os.path.join(dir_name, ref_start_filename))
ref_start_gray = cv2.cvtColor(ref_start, cv2.COLOR_BGR2GRAY)
misc.toimage(ref_start_gray)
```

```python
# Loading the reference markers
ref_end = cv2.imread(os.path.join(dir_name, ref_end_filename))
ref_end_gray = cv2.cvtColor(ref_end, cv2.COLOR_BGR2GRAY)
misc.toimage(ref_end_gray)
```

```python
# Finding marker positions in the incoming image: Start
small_image = ref_start_gray
large_image = img_form_gray
method = cv2.TM_SQDIFF_NORMED
result = cv2.matchTemplate(small_image, large_image, method)
_,_,s_mnLoc,_ = cv2.minMaxLoc(result)
```

```python
# Finding marker positions in the incoming image: End
small_image = ref_end_gray
large_image = img_form_gray
method = cv2.TM_SQDIFF_NORMED
result = cv2.matchTemplate(small_image, large_image, method)
_,_,e_mnLoc,_ = cv2.minMaxLoc(result)
```
