## import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import misc

```python
img1 = cv2.imread('samples/Claim_SamGray2.jpg',0)
img2 = cv2.imread('samples/CMS1500_FrasierClean.jpg',0)
```

```python
img1_invert = cv2.bitwise_not(img1)
img1_blur = cv2.GaussianBlur(img1_invert, (5, 5), 0)
img1_ret, img1_thresh = cv2.threshold(img1_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# plt.imshow(img1_thresh)
# plt.show()
```

```python
img2_invert = cv2.bitwise_not(img2)
img2_blur = cv2.GaussianBlur(img2_invert, (5, 5), 0)
img2_ret, img2_thresh = cv2.threshold(img2_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# plt.imshow(img2_thresh)
# plt.show()
```

```python
# This is the detector we're going to use for the features.
orb = cv2.ORB_create()
```

```python
# Here, we find the key points and their descriptors with the orb detector.
kp1, des1 = orb.detectAndCompute(img1_thresh,None)
kp2, des2 = orb.detectAndCompute(img2_thresh,None)
```

```python
# This is our BFMatcher object.
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
# Here we create matches of the descriptors, then we sort them based on their distances
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
# Here, we've drawn the first 10 matches
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:3],None, flags=2)
plt.imshow(img3)
plt.show()
```

```python
misc.imsave('matched.jpg', img3)
```
