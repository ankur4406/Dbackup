```python
import cv2
from scipy import misc
```

```python
omr_file = r'slices\BillingContactInfo_AffliatedWithAnExistingContract_slice.png'
omr_slice = cv2.imread(omr_file)
misc.toimage(omr_slice)
```

```python
#convert to gray scale
gray_image = cv2.cvtColor(omr_slice, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
thresh
```

```python
misc.toimage(thresh)
```

```python
misc.toimage(omr_slice)
```

```python
image, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img = omr_slice.copy()
img = cv2.drawContours(img, contours, -1, (255, 0, 255), 4)
cv2.imwrite('image_contours.png', img)
```
