```python
import os
import cv2
# Loading the incoming CMS1500 Form in the form of an array
img1 = cv2.imread(os.path.join(dir_name, "references\\element_marker_2.png"))
img2 = cv2.imread(os.path.join(dir_name, in_form))
```

```python
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray = cv2.bitwise_not(gray)
plt.imshow(gray),plt.show()
```

```python
thresh = cv2.threshold(gray, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
# plt.imshow(thresh),plt.show()
misc.toimage(img1)
```

```python
misc.toimage(thresh)
```

```python
import numpy as np
from matplotlib import pyplot as plt

# img1 = cv2.imread('box.png',0)          # queryImage
# img2 = cv2.imread('box_in_scene.png',0) # trainImage

# Initiate SIFT detector
orb = cv2.ORB_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(thresh,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
kp1
```

```python
# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
# Match descriptors.
matches = bf.match(des1,des2)
```

```python
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 10 matches.
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:], None, flags=2)

plt.figure(figsize=(30,20))
plt.imshow(img3),plt.show()
```

```python
img = img2

# Initiate STAR detector
orb = cv2.ORB_create()

# find the keypoints with ORB
kp = orb.detect(img,None)

# compute the descriptors with ORB
kp, des = orb.compute(img, kp)

# draw only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(img,kp,None, color=(0,255,0), flags=0)

plt.figure(figsize=(30,20))
plt.imshow(img2),plt.show()
```
